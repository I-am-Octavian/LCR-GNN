```python
import os
import spacy
import json
from data_preparation import get_json_from_spacy_doc
from legal_ner import extract_entities_from_judgment_text
```

```python
import torch
print(torch.cuda.is_available())
spacy.require_gpu(0)
```

```python

legal_nlp = spacy.load('en_legal_ner_trf')

preamble_spiltting_nlp = spacy.load('en_core_web_sm')
```

```python

input_dir = '../DATASET/judgement' # Replace with your input directory
output_dir = '../DATASET/Relations' # Replace with your output directory

if not os.path.exists(output_dir):
    os.makedirs(output_dir)

print(len(os.listdir(input_dir)))

run_type='sent'
# Get all txt files
txt_files = [f for f in os.listdir(input_dir) if f.endswith(".txt")]

print(txt_files[0])
```

```python
for filename in txt_files:
  output_filename = os.path.splitext(filename)[0] + '.json'
  if os.path.isfile(os.path.join(output_dir, output_filename)):
    print(f'Skipping{output_filename}')
  else:
    with open(os.path.join(input_dir, filename), 'r') as f:
        judgment_text = f.read()
    
    try:
      combined_doc = extract_entities_from_judgment_text(judgment_text, legal_nlp, preamble_spiltting_nlp, run_type, do_postprocess=True)
    except RuntimeError as e:
      torch.cuda.empty_cache()
      print(f'Error occured for file: {filename}')
      continue

    with open(os.path.join(output_dir, output_filename), 'w') as f:
        json.dump(get_json_from_spacy_doc(combined_doc), f, indent=4)
```
